```
import pandas as pd
ratings1=pd.read_csv("imdb.csv", error_bad_lines=False)
ratings1.head()
```

```
ratings1.describe()
```

```
C= ratings1['imdbRating'].mean()
C
```

```
m= ratings1['ratingCount'].quantile(0.9)
m
```

```
movies = ratings1.copy().loc[ratings1['ratingCount'] >= m]
movies.shape
```

```
def rating_score(x, m=m, C=C):
    v = x['ratingCount']
    R = x['imdbRating']
    # Calculation by using IMDB formula
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```
movies['score'] = movies.apply(rating_score, axis=1)
```

```
movies = movies.sort_values('score', ascending=False)
```

```
movies[['title', 'ratingCount', 'imdbRating', 'score']].head(10)
```

```
trending=[i for i in range(1,(movies.shape[0]+1))]
movies["Trending"]=trending
movies[['Trending','title', 'ratingCount', 'imdbRating', 'score']].head(10)
```

```
In this way we create a new column called Trending which recommends the movies according to trends. 
```
